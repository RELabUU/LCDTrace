```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Create Subplot

```python
# set Colors
color1 = '#ec5f66'
color2 = '#66ec5f'
color3 = '#5f66ec'
edgeColor = 'white'


#Set Labels
labels = ['P', 'R', 'F0.5', 'F1', 'F2']

# set width of bars
barWidth = 0.2

# Set position of bar on X axis
r1 = np.arange(5)
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

#Create Sub Plot
fig, axes = plt.subplots(nrows=3, ncols=4, sharey=True)
fig.set_size_inches(20, 20)

#naming cols and rows
rows = ['Project 2', 'Project 3', 'Project 4']
cols = ['None', 'SMOTE', 'Undersampling', '5050']


for ax, col in zip(axes[0], cols):
    ax.set_title(col)

for ax, row in zip(axes[:,0], rows):
    ax.set_ylabel(row, rotation=90, size='large')

```
